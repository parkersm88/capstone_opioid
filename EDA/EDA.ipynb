```python
calls=data[['Month',"TotalNarcan",'MPDCalls', 'FEMSCalls','MedicalCalls', 'MiscCals', 'OverdoseCalls', 'PropertyCalls',  'QOLCalls', 'TrafficCalls', 'ViolentCalls']]
calls.head()      
     
```

```python
call_group=calls.groupby('Month').sum().reset_index()
call_group
```

```python
import seaborn as sns
# Red dot with Line
fig, ax = plt.subplots()

ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)  


ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left() 
plt.xticks(fontsize=12) 

ax.plot(call_group['Month'], call_group['OverdoseCalls'],'k')
ax.plot(call_group['Month'], call_group['OverdoseCalls'],'ro')


ax.set_title("OverdoseCalls")
plt.show()

```

```python
import matplotlib.lines as mlines
fig, ax = plt.subplots()

ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)  


ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left() 
plt.xticks(fontsize=12) 

ax.plot(call_group['Month'], call_group['OverdoseCalls'],'k')
ax.plot(call_group['Month'], call_group['OverdoseCalls'],'ro')

ax.plot(call_group['Month'], call_group['TotalNarcan'],'g')
ax.plot(call_group['Month'], call_group['TotalNarcan'],'bo')

#Legend Stuff
green_line = mlines.Line2D([], [], color='green', marker='o',markerfacecolor='blue',
                          markersize=7, label='TotalNarcan')
black_line = mlines.Line2D([], [], color='black', marker='o',markerfacecolor='darkred',
                          markersize=7, label='OverdoseCalls')

ax.legend(handles=[green_line,black_line], loc='best')



ax.set_title("OverdoseCalls vs TotalNarcan")
plt.show()
```
